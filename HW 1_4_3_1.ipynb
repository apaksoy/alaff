```python
# Homework 1.1.1.4 as well

import numpy as np
from scipy import linalg

np.random.seed(0)

n = 3
x = n * np.random.random((n, 1))
U = n * np.triu(np.random.random((n, n)), 0)
print(f"{x = }\n")
print(f"{U = }\n")

b = U @ x
print(f"{b = }\n")

x_hat = linalg.solve(U,b)

print(f"{np.linalg.norm(x_hat - x) = }\n")
print(f"{np.linalg.norm(b - U @ x_hat) = }\n")
     
print(f"{min(np.diag(U)) = }\n")

print(f"{np.linalg.cond(U) = }\n")
```

```python
# Extension of HW 1.4.3.1
# What happens as the matrix size grows larger?

import numpy as np
from scipy import linalg
import matplotlib.pyplot as plt

np.random.seed(2)

n = 1000
'''
A = 1 * np.random.random((1, 1))
U = np.triu(A, 0)
cond_nums = np.array([[1, np.linalg.cond(A), np.linalg.cond(U)]])
'''
# matrix number for a scalar is always 1, 
cond_nums = np.array([[1,1,1]]) 
# starting from 2 because np.append() can not append an empty matrix
for i in range(10,1001,10): 
    A = np.random.random((i, i)) 
    U = np.triu(A, 0)
    cond_nums = np.append(cond_nums,[[i, np.linalg.cond(A,2), np.linalg.cond(U,2)]], axis = 0)
   
fig, ax = plt.subplots()
fig.suptitle("The 2-norm condition number vs size of random matrices")
plt.xlabel('Matrix size')
plt.ylabel('Condition number (logscale)')
plt.yscale("log")
ax.plot(cond_nums[:,0], cond_nums[:,1], label = "regular")
ax.plot(cond_nums[:,0], cond_nums[:,2], label = "upper triangular")
plt.legend()
plt.show()
```
